```python
# 插入排序
def insertSort(mylist):
    for i in range(len(mylist)):
        if i == 0:
            continue
        for j in range(0, i):
            if mylist[i - j] < mylist[i - j - 1]:
                mylist[i - j - 1], mylist[i - j] = mylist[i - j], mylist[i - j - 1]


# 冒泡排序
def bubblesort(mylist):
    for i in range(len(mylist), 0, -1):
        if i == 0: continue
        for j in range(0, i):
            if j != i - 1 and mylist[j] > mylist[j + 1]:
                mylist[j], mylist[j + 1] = mylist[j + 1], mylist[j]


# 选择排序
def selectsort(mylist):
    def findmin(l, a):
        minitem = a
        for f in range(a, len(l)):
            if l[f] < l[minitem]:
                minitem = f
        return minitem

    for i in range(0, len(mylist)):
        mini = findmin(mylist, i)
        if mylist[i] != mylist[mini]:
            mylist[i], mylist[mini] = mylist[mini], mylist[i]


# 二路归并
def mergesort(mylist):
    def merge(left, right):
        merged = []
        i = j = 0
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[j])
                j += 1
        merged.extend(left[i:])
        merged.extend(right[j:])
        return merged

    if len(mylist) <= 1:
        return mylist
    mid = len(mylist) // 2
    left = mylist[:mid]
    right = mylist[mid:]
    return merge(mergesort(left), mergesort(right))


# 快速排序
def quicksort(mylist):
    if len(mylist) <= 1:
        return mylist
    pivot = mylist[0]
    left = [x for x in mylist[1:] if x < pivot]
    right = [x for x in mylist[1:] if x >= pivot]
    return quicksort(left) + [pivot] + quicksort(right)


# 基数排序
def cardinalSorting(mylist):
    def getDigit(x, d):
        a = [int(i) for i in str(x)]
        a.reverse()
        if d > len(a):
            return 0
        else:
            return a[d - 1]

    def radixsort(mylist, d):
        for i in range(1, d + 1):
            s = [[] for x in range(10)]
            for x in mylist:
                s[getDigit(x, i)].append(x)
            mylist = [x for y in s for x in y]
        return mylist

    d = len(str(max(mylist)))
    return radixsort(mylist, d)


# 求第k小的数
def selectKth(mylist, k):
    pivot = mylist[0]
    left = [x for x in mylist[1:] if x < pivot]
    right = [x for x in mylist[1:] if x >= pivot]
    if len(left) == k - 1:
        return pivot
    elif len(left) > k - 1:
        return selectKth(left, k)
    else:
        return selectKth(right, k - len(left) - 1)


mylist = [38, 27, 43, 3, 9, 82, 10]
sorted_list = mergesort(mylist)
print(sorted_list)
```
